```python
missing_threshold = 0.3
df.dropna(thresh=len(df) * (1 - missing_threshold), axis=1, inplace=True)
```

```python
categorical_columns = ['region_txt', 'attacktype1_txt', 'targtype1_txt'] 
df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
for column in df.columns:
    if df[column].isnull().mean() > 0.3:
        df[column].fillna(df[column].mean(), inplace=True)
```

DATA VISUALIZATION

```python
# The country that faced the most attacks each year.
attacks_per_year_country = df.groupby(['iyear', 'country_txt']).size().reset_index(name='attack_count')
most_attacked_per_year = attacks_per_year_country.groupby('iyear')['attack_count'].idxmax()
most_attacked_countries = attacks_per_year_country.loc[most_attacked_per_year]
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='iyear', y='attack_count', hue='country_txt', data=most_attacked_countries)
plt.title('Country that Faced the Most Attacks per Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#Number of attacks per region with success rate
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='region_txt', hue='success', palette='Set2')
plt.title('Number of Attacks per Region (Hue: Success)')
plt.xlabel('Region')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.legend(title='Success', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
df['region_txt'].value_counts()
```

```python
#Pie chart of terrorist attack types with percentage
attack = df['attacktype1_txt'].value_counts().sort_values(ascending=False)
plt.pie(attack.values, labels=attack.index, autopct="%1.1f%%")
plt.title("Types of Terrorist Attacks")
plt.show()
```

```python
#Attack types with their total number
plt.rcParams['figure.figsize']=15,11
sns.barplot(x=df['attacktype1_txt'].value_counts()[:20].index,y=df['attacktype1_txt'].value_counts()[:20].values)
plt.ylabel('Total Attack', fontsize=20)
plt.xticks(rotation=90)
plt.xlabel('Attack Type', fontsize=20)
plt.title('Attack types',size=25, fontweight="bold")
```

```python
#Target type vs Attack Type
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='targtype1_txt', hue='attacktype1_txt', palette='Set2')
plt.title('Target Type vs. Attack Type')
plt.xlabel('Target Type')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Attack Type', loc='upper right', bbox_to_anchor=(1.2, 1))
plt.tight_layout()
plt.show()
```

```python
attacks_per_year = df.groupby('iyear').size()
plt.figure(figsize=(12, 6))
plt.plot(attacks_per_year.index, attacks_per_year.values, marker='o', linestyle='-')
plt.title('Number of Attacks Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.grid(True)
plt.show()
```

```python
correlation_matrix = df.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.tight_layout()
plt.show()
```

```python
deaths_by_country = df.groupby('country_txt')['success'].sum()

plt.figure(figsize=(12, 6))
plt.bar(deaths_by_country.index[:10], deaths_by_country.values[:10])
plt.xlabel("Country")
plt.ylabel("Number of Deaths")
plt.title("Countries with Most Deaths from Terrorism")
plt.xticks(rotation=90)
plt.show()
```

```python
deaths_by_attacktype = df.groupby('attacktype1_txt')['success'].sum()

plt.figure(figsize=(12, 6))
plt.bar(deaths_by_attacktype.index, deaths_by_attacktype.values)
plt.xlabel("Attack Type")
plt.ylabel("Number of Deaths")
plt.title("Number of Deaths from Different Attack Types")
plt.xticks(rotation=90)
plt.show()
```
